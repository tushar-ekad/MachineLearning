```python
import pandas as pd
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
titanic_df = pd.read_csv('train.csv')
titanic_df.sample(7)
```

```python
flights_df = sns.load_dataset('flights')
flights_df.sample(7)
```

```python
iris_df = sns.load_dataset('iris')
iris_df.sample(7)
```

## 1. Scatterplot (Numerical-Numerical)

```python
sns.scatterplot(x= tips_df['total_bill'], y=tips_df['tip'])
```

```python
sns.scatterplot(x= tips_df['total_bill'], y=tips_df['tip'], hue=tips_df['sex'], style=tips_df['smoker'], size=tips_df['size'])
```

## 2. Bar Plot(Numerical-Categorical)

```python
sns.barplot(x=titanic_df['Pclass'], y=titanic_df['Age'], hue=titanic_df['Sex'])
```

```python
sns.barplot(x=titanic_df['Pclass'], y=titanic_df['Fare'], hue=titanic_df['Sex'])
```

## 3. Box Plot (Numerical-Categorical)

```python
sns.boxplot(x=titanic_df['Sex'], y=titanic_df['Age'], hue=titanic_df['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic_df[titanic_df['Survived'] == 0]['Age'], hist=False)
sns.distplot(titanic_df[titanic_df['Survived'] == 1]['Age'], hist=False)
```

## 5. HeatMap (Categorical - Categorical)

```python
pd.crosstab(titanic_df['Pclass'], titanic_df['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic_df['Pclass'], titanic_df['Survived']))
```

```python
titanic_df.groupby('Pclass')['Survived'].mean()*100
```

```python
titanic_df.groupby('Embarked')['Survived'].mean()*100
```

```python
(titanic_df.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

## 6. ClusterMat (Categorical - Categorical)

```python
pd.crosstab(titanic_df['SibSp'], titanic_df['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic_df['SibSp'], titanic_df['Survived']))
```

## Pairplot

```python
sns.pairplot(iris_df, hue='species')
```

```python
sns.pairplot(titanic_df, hue='Sex')
```

## 8. Lineplot (Numerical - Numerical)

```python
new = flights_df.groupby('year').sum(numeric_only=True).reset_index()
```

```python
sns.lineplot(x= new['year'], y=new['passengers'])
```

```python
flights_df.pivot_table(values='passengers', index='month', columns='year')
```

```python
sns.heatmap(flights_df.pivot_table(values='passengers', index='month', columns='year'))
```

```python
sns.clustermap(flights_df.pivot_table(values='passengers', index='month', columns='year'))
```
