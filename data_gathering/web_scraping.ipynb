```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}

webpage  = requests.get('https://www.ambitionbox.com/list-of-companies?page=1', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_='companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []
review_count = []
desc_loc = []
highly_rated_for = []
critically_rated_for = []
salaries = []
interviews = []
jobs = []
benefits = []
photos = []

for i in company:
    name.append(i.find('h2', class_='companyCardWrapper__companyName').text.strip())
    rating.append(i.find('div', class_='rating_text').text.strip())
    review_count.append(i.find('span', class_='companyCardWrapper__companyRatingCount').text.strip().strip('()'))
    desc_loc.append(i.find('span', class_ = 'companyCardWrapper__interLinking').text.strip())
    high = 'None'
    low = 'None'

    headers = i.find_all('div', class_='companyCardWrapper__ratingHeader')

    for header in headers:

        text = header.text.strip()

        value_tag = header.find_next(
            'span',
            class_='companyCardWrapper__ratingValues'
        )

        value = value_tag.text.strip() if value_tag else 'None'

        if text == 'Highly Rated For':
            high = value

        elif text == 'Critically Rated For':
            low = value

    highly_rated_for.append(high)
    critically_rated_for.append(low)

    salaries.append(i.find_all('span', class_ = 'companyCardWrapper__ActionCount')[1].text.strip())
    interviews.append(i.find_all('span', class_ = 'companyCardWrapper__ActionCount')[2].text.strip())
    jobs.append(i.find_all('span', class_ = 'companyCardWrapper__ActionCount')[3].text.strip())
    benefits.append(i.find_all('span', class_ = 'companyCardWrapper__ActionCount')[4].text.strip())
    photos.append(i.find_all('span', class_ = 'companyCardWrapper__ActionCount')[5].text.strip())
```

```python
print(len(name))
print(len(rating))
print(len(review_count))
print(len(desc_loc))
print(len(highly_rated_for))
print(len(critically_rated_for))
print(len(salaries))
print(len(interviews))
print(len(jobs))
print(len(benefits))
print(len(photos))
```

```python
data = {
    "name" : name,
    "rating" : rating,
    "review_count" : review_count,
    "desc_loc" : desc_loc,
    "highly_rated_for" : highly_rated_for,
    "critically_rated_for" : critically_rated_for,
    "salaries" : salaries,
    "interviews" : interviews,
    "jobs" : jobs,
    "benefits" : benefits,
    "photos" : photos,
}

df = pd.DataFrame(data=data)

df
```

## scrap 500 pages

```python
name = []
rating = []
review_count = []
desc_loc = []
highly_rated_for = []
critically_rated_for = []
salaries = []
interviews = []
jobs = []
benefits = []
photos = []

headers = {
    "User-Agent": "Mozilla/5.0"
}

session = requests.Session()

for page in range(1, 501):

    url = f'https://www.ambitionbox.com/list-of-companies?page={page}'

    response = session.get(url, headers=headers)

    soup = BeautifulSoup(response.text, 'lxml')

    companies = soup.find_all('div', class_='companyCardWrapper')

    for company in companies:

        # Basic details
        name.append(
            company.find('h2', class_='companyCardWrapper__companyName')
            .text.strip()
        )

        rating_tag = company.find('div', class_='rating_text')
        rating.append(rating_tag.text.strip() if rating_tag else '-')

        review_count_tag = company.find('span', class_='companyCardWrapper__companyRatingCount')
        review_count.append(review_count_tag.text.strip().strip('()')) if review_count_tag else '-'

        desc_loc_tag = company.find('span', class_='companyCardWrapper__interLinking')
        desc_loc.append(desc_loc_tag.text.strip().strip('()')) if desc_loc_tag else '-'

        # High / Low ratings
        high = 'None'
        low = 'None'

        rating_headers = company.find_all(
            'div',
            class_='companyCardWrapper__ratingHeader'
        )

        for header in rating_headers:

            title = header.text.strip()

            value_tag = header.find_next(
                'span',
                class_='companyCardWrapper__ratingValues'
            )

            value = value_tag.text.strip() if value_tag else 'None'

            if title == 'Highly Rated For':
                high = value

            elif title == 'Critically Rated For':
                low = value

        highly_rated_for.append(high)
        critically_rated_for.append(low)

        # Action counts
        counts = company.find_all(
            'span',
            class_='companyCardWrapper__ActionCount'
        )

        salaries.append(counts[1].text.strip() if len(counts) > 1 else 'None')
        interviews.append(counts[2].text.strip() if len(counts) > 2 else 'None')
        jobs.append(counts[3].text.strip() if len(counts) > 3 else 'None')
        benefits.append(counts[4].text.strip() if len(counts) > 4 else 'None')
        photos.append(counts[5].text.strip() if len(counts) > 5 else 'None')

    print(f'Page {page} done')
```

```python
data = {
    "name" : name,
    "rating" : rating,
    "review_count" : review_count,
    "desc_loc" : desc_loc,
    "highly_rated_for" : highly_rated_for,
    "critically_rated_for" : critically_rated_for,
    "salaries" : salaries,
    "interviews" : interviews,
    "jobs" : jobs,
    "benefits" : benefits,
    "photos" : photos,
}

df = pd.DataFrame(data=data)
```

```python
df.to_csv('../data/data_gathering/ambition_box_companies.csv')
```
