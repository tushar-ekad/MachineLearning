## Opening a csv file from URL

```python
## To read csv from a url we can use the same function but with the url as an argument
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url=url, headers=headers)
data = StringIO(req.text)
pd.read_csv(data)
```

## Sep parameter
    - By default, pandas assumes that the separator in a csv file is a comma. If the separator is something else, we can specify it using the sep parameter.
    - If the dataset don't have headers/column names we can add those using 'names' parameter


```python
# pd.read_csv('../data/working_with_csv/movie_titles_metadata.tsv')
# pd.read_csv('../data/working_with_csv/movie_titles_metadata.tsv', sep='\t')
pd.read_csv('../data/working_with_csv/movie_titles_metadata.tsv', sep='\t', names=['sno', 'name', 'release_year', 'rating', 'votes', 'genres'])
```

## index_col parameter
    - To set a column as index we can use index_col parameter

```python
# pd.read_csv('../data/working_with_csv/aug_train.csv')
pd.read_csv('../data/working_with_csv/aug_train.csv', index_col='enrollee_id')
```

## Header parameter
    - if the csv file has some extra rows at the top which are not part of the data we can use header argument to specify the row number which contains the column names

```python
# pd.read_csv('../data/working_with_csv/test.csv')
pd.read_csv('../data/working_with_csv/test.csv', header=1)
```

## use_cols parameter
    - To read only specific columns from a csv file we can use the usecols parameter

```python
# pd.read_csv('../data/working_with_csv/aug_train.csv')
pd.read_csv('../data/working_with_csv/aug_train.csv', usecols=['enrollee_id', 'gender', 'education_level'])
```

## skiprows and nrows parameter
    - To read only specific rows from a csv file we can use the skiprows and to read only specific number of rows we can use nrows

```python
pd.read_csv('../data/working_with_csv/aug_train.csv', skiprows=[0,3], nrows=10)
```

## encoding parameter
    - To read csv with different encoding we can use the encoding parameter

```python
pd.read_csv('../data/working_with_csv/zomato.csv', encoding='latin-1')
```

## skip bad lines
    - to skip bad lines(rows where some values are empty) we can use on_bad_lines parameter

```python
pd.read_csv('../data/working_with_csv/BX-Books.csv', sep=';', encoding='latin-1', on_bad_lines='skip')
```

## dtypes parameter
    - to read a csv file with specific data types for columns we can use dtype parameter

```python
# pd.read_csv('../data/working_with_csv/aug_train.csv').info()
# pd.read_csv('../data/working_with_csv/aug_train.csv', dtype={'target':int})
pd.read_csv('../data/working_with_csv/aug_train.csv', dtype={'target':int}).info()
```

## handling dates

```python
pd.read_csv('../data/working_with_csv/IPL Matches 2008-2020.csv', parse_dates=['date']).info()
# combine date and time columns to create a datetime column
# pd.read_csv('../data/working_with_csv/IPL Matches 2008-2020.csv', parse_dates=[['date', 'time']]).info()
```

## convertors
    - to apply a function to a column while reading a csv file we can use converters parameter

```python
pd.read_csv('../data/working_with_csv/IPL Matches 2008-2020.csv')
```

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else:
        return name
    
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('../data/working_with_csv/IPL Matches 2008-2020.csv', converters={'team1':rename})
```

## na_values parameter
    - to specify values to be considered as NaN while reading a csv file we can use na_values parameter useful like when values are missing but represented as some string like '-'

```python
# pd.read_csv('../data/working_with_csv/aug_train.csv')
pd.read_csv('../data/working_with_csv/aug_train.csv', na_values=['Male'])
```

## Loading a huge dataset in chunks
    - To read a huge csv file we can use chunksize parameter to read the file in chunks and avoid memory error. and then we can do our task with loop

```python
dfs = pd.read_csv('../data/working_with_csv/aug_train.csv', chunksize=1000)
```

```python
for chunk in dfs:
    print(chunk.shape)
```
